```python
import pandas as pd
import datetime 
```

#### Cleaning up February 2021 data for use in Tableau

```python
raw_bike21_df = pd.read_csv("./raw_data/202102-citibike-tripdata.csv")
# raw_bike21_df
```

```python
# raw_bike21_df.isnull().sum()
```

```python
# converting day time column from object to datetime type
raw_bike21_df['dt_starttime'] = pd.to_datetime(raw_bike21_df['starttime'])
# raw_bike21_df.head()
```

```python
#creating columns of only time and date for possible use in Tableau
raw_bike21_df['dt_start_date'] = raw_bike21_df['dt_starttime'].dt.date
raw_bike21_df['dt_starting_time'] = raw_bike21_df['dt_starttime'].dt.time
# raw_bike21_df.head()
```

```python
# setting dt_starttime as index to allow for filtering using datetime function
index_bike21_df = raw_bike21_df.set_index('dt_starttime')
# index_bike21_df.head(2)
```

```python
# create new dfs of activity designated ranges for February 2021
am_target_bike21_df = index_bike21_df.between_time('00:00', '06:00')
morn_target_bike21_df = index_bike21_df.between_time('06:01', '12:00')
day_target_bike21_df = index_bike21_df.between_time('12:01', '18:00')
pm_target_bike21_df = index_bike21_df.between_time('18:01', '23:59')

# reset index and then move dt_starttime to end of data set
am_bike21_df = am_target_bike21_df.reset_index()
morn_bike21_df = morn_target_bike21_df.reset_index()
day_bike21_df = day_target_bike21_df.reset_index()
pm_bike21_df = pm_target_bike21_df.reset_index()

# removing (pop) dt_starttime column so it can be replaced at at the end
am_datetime21 = am_bike21_df.pop('dt_starttime')
morn_datetime21 = morn_bike21_df.pop('dt_starttime')
day_datetime21 = day_bike21_df.pop('dt_starttime')
pm_datetime21 = pm_bike21_df.pop('dt_starttime')

#re-inserting column at end of df 
am_bike21_df['dt_starttime'] =  am_datetime21
morn_bike21_df['dt_starttime'] = morn_datetime21
day_bike21_df['dt_starttime'] =  day_datetime21
pm_bike21_df['dt_starttime'] =  pm_datetime21

# pm_bike21_df
```
